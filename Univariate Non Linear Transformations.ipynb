```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split

rnd = np.random.RandomState(0)
# print(rnd)
X_org = rnd.normal(size=(1000, 3))
# print(X_org.shape)
# print(X_org)
w = rnd.normal(size=3)
# print(w)
X = rnd.poisson(10 * np.exp(X_org))
y = np.dot(X_org, w)
print("Number of feature appearances:{}".format(np.bincount(X[:, 0])))

# bins = np.bincount(X[:, 0])
# plt.bar(range(len(bins)), bins, color='grey')
# plt.xlabel("Features")
# plt.ylabel("Number of appearences")

#trying to fit a linear model
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
print("score on ridge linear model:{}".format(Ridge().fit(X_train, y_train).score(X_test, y_test)))

X_train_log = np.log(X_train+1)
X_test_log = np.log(X_test+1)

plt.hist(np.log(X_train_log[:, 0]+1), bins=25, color='grey')
plt.xlabel("Values")
plt.ylabel("Number of appearences")

print("score after using function log:{}".format(Ridge().fit(X_train_log, y_train).score(X_test_log, y_test)))
```
