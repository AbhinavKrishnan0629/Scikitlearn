```python
#use of binning

import mglearn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder

X, Y = mglearn.datasets.make_wave(n_samples=100)
line = np.linspace(-3, 3, 1000, endpoint=True).reshape(-1, 1)

bins = np.linspace(-3, 3, 11)
which_bin = np.digitize(X, bins=bins)
encoder = OneHotEncoder(sparse=False)
encoder.fit(which_bin)
X_binned = encoder.transform(which_bin)

which_line = np.digitize(line, bins=bins)
encoder.fit(which_line)
line_binned = encoder.transform(which_line)

reg1 = DecisionTreeRegressor(min_samples_split=3).fit(X_binned, Y)
reg2 = LinearRegression().fit(X_binned, Y)

plt.plot(line, reg1.predict(line_binned), label='D')
plt.plot(line, reg2.predict(line_binned), label='l')
plt.plot(X[:, 0], Y, 'o', c='k')
plt.legend(loc='best')
```
