```python
import mglearn
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs

X, y = make_blobs(random_state=0, n_samples=12)

dbscan = DBSCAN()
clusters = dbscan.fit_predict(X)

print("cluster membership{}".format(clusters))
#all the points are labelled as outliers or not given a cluster because of eps and min_samples parameter are not tuned properly for the toy datasets
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

X, y = make_moons(random_state=0, n_samples=200, noise=0.05)

scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)

dbscan = DBSCAN()
clusters = dbscan.fit_predict(X_scaled)

print("Clusters:{}".format(clusters))

plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap=mglearn.cm3)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
```
